```python
import pandas as pd ; import numpy as np
```

```python
vs=pd.DataFrame({"anahtar":list("ABC")*2,
                "veri1":range(6),
                "veri2":np.arange(5,11)})
vs
```

```python
grup=vs.groupby("anahtar")
```

```python
grup.aggregate(["min",np.median,"max"])
# birden fazla işlem yapmak için aggergate kullanıyoruz
```

```python
    grup.agg({"veri1":"min","veri2":"max"})
```

```python
def f(x):
    return x.max()-x.min()
# veri 1 için max-min yaptı altta da yazdırdık
# aynısını veri 2 içinde yaptı
```

```python
grup.agg(f)
```

```python
veri=pd.DataFrame({"harf":list("ABC")*4,
                  "sayi":["bir","iki"]*6,
                  "v1":np.random.randn(12),
                  "v2":np.arange(10,33,2)})
veri
```

```python
grup=veri.groupby(["harf","sayi"])
```

```python
grup_v1=grup["v1"]
# v1 sütununu grup_v1 e aktardı
```

```python
grup_v1.agg("mean")
# burda da ortalamasını aldı
```

```python
grup_v1.agg(["mean","std",f])
```

```python
grup_v1.agg([("falan","mean"),("filan",np.std)])
# mean falan olarak görünüyor tablo da 
# filan kısmı da std kısımları görünüyor
```

```python
fonk=["mean","count","max"]
result=grup[["v1","v2"]].agg(fonk)
result
```
