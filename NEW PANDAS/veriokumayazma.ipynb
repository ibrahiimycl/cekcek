```python
import pandas as pd
```

```python
df=pd.read_table("veri1.txt")
df
```

```python
df = pd.read_table('veri1.txt', sep='\s+')
df
# sep='\s+', pd.read_table() fonksiyonunda kullanılan bir parametredir 
# ve dosyanın sütunları arasında ayırıcı olarak hangi karakterin kullanılacağını belirtir. 
# \s+, bir veya daha fazla boşluk karakteri (boşluk, sekme, satır başı vb.) anlamına gelir. 
# Yani, sep='\s+' parametresi, dosyanın sütunlarını boşluk karakterleriyle ayıracağını belirtir.
```

```python
df=pd.read_table("veri2.txt")
df
```

```python
df=pd.read_table("veri2.txt", sep=",")
df
# sep ile virgülleri kaldırdık
```

```python
df=pd.read_table("veri2.txt", sep=",", header=None)
df
# header=None sütun yok dedik o da sütun yaptı bize
# header=None parametresini kullanarak
# dosyanın ilk satırının sütun isimlerini içermediğini belirtebilirsiniz.
# Bu parametre dosyayı okurken ilk satırı sütun isimleri olarak almaz
# sütunları varsayılan bir numaralandırmayla oluşturur.
```

```python
df=pd.read_table("veri2.txt", sep=",", header=None,
                names=["isim","puan","cinsiyet"])
df
# names yerine başka birşey yazamayız
# çünkü names özel bir fonksiyondur
```

```python
df=pd.read_table("veri2.txt", sep=",", header=None,
                names=["isim","puan","cinsiyet"],
                index_col="isim")
df
# index_col özel parametredir istediğimi sütundaki yazıyı 
# index olarak yazdırır
```

```python
df=pd.read_table("veri4.txt")
df
```

```python
df=pd.read_table("veri4.txt", sep=",")
df
```

```python
df=pd.read_table("veri4.txt", sep=",", skiprows=[0,2])
df
# skiprowsun içine okutmak istemediğimiz 
# satırların indexini yazınca oraları almıyor
```

```python
df=pd.read_table("veri4.txt", sep=",", skiprows=[0,2],
                usecols=[0,1])
df
# usecolsun içine okutmak istediğimiz sütunların
# indexini yazıyoruz
```

```python
df=pd.read_table("veri4.txt", sep=",", skiprows=[0,2],
                usecols=[0,1], nrows=2)
df
# nrows bize okutmak istediğimiz satırları göstermek içindir
```

```python
df=pd.read_table("veri5.txt", sep=",")
df
# tabloda eksik veriler NAN olarak görünüyor
```

```python
pd.isnull(df)
# true yazan kısımlarda eksik veri olduğu söyleniyor
```

```python
df=pd.read_table("veri5.txt", sep=",", na_values=["n.a",-5])
df
# na_valuesın içine tabloda neleri eksik veri almak istiyorsak
# onu yazıyoruz o da onu NAN olarak gösteriyor
```

```python
eksik_veri={"puan":[-5],"cinsiyet":["efe","n.a"]}
df=pd.read_table("veri5.txt", sep=",", na_values=eksik_veri)
df
# burada neden eksik veri kısmını ekledik?
# çünkü hem cinsiyette hemde isimde efe yazıyor
# sadece cinsiyetteki efeyi eksik veri olarak yazdırmak için
# bu yöntemi kullandık tabi diğer bilgileri de ekledik
```
