```python
iris=pd.read_csv("iris.txt", sep="\t")
iris.head()
```

```python
iris.plot.hist(alpha=0.9)
```

```python
iris.plot.hist(alpha=0.9,stacked=True)
```

```python
iris.plot.hist(alpha=1,stacked=True,bins=20)
# iris: Bu, histogramlarını oluşturacağımız veri kümesidir. Önceden yüklenmiş olması gerekmektedir.
# plot.hist(): Bu, veri kümesindeki özelliklerin histogramlarını oluşturmak için kullanılan bir fonksiyondur.
# alpha=1: Histogramların opaklığını belirler. Burada 1 değeri, tamamen opak histogramlar elde edeceğimizi belirtir.
# stacked=True: Histogramların yığınlanıp yığınlanmayacağını belirler.
# True olduğunda, histogramlar birbirinin üzerine yığılarak gösterilir.
# Bu, farklı özelliklerin dağılımlarını karşılaştırmak için kullanışlı olabilir.
# bins=20: Histogram çubuklarının sayısını belirtir. Bu durumda, her özellik için 20 histogram çubuğu oluşturulur.
```

```python
iris["Canak_yaprak_eni"].plot.hist(orientation="horizontal")
```

```python
iris["Tac_yaprak_boyu"].diff().hist()
# iris["Tac_yaprak_boyu"]: "iris" veri kümesindeki "Tac_yaprak_boyu" adlı sütunu seçer.
# diff(): Seçilen sütundaki ardışık değerler arasındaki farkı hesaplar.
# Bu, veri kümesindeki ardışık ölçümler arasındaki değişimi hesaplamak için kullanılır.
# hist(): Hesaplanan farkların histogramını çizer.
```
