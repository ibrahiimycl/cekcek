```python
import numpy as np ; import pandas as pd;
```

```python
v1=pd.DataFrame({"anahtar":["a","b","c","c","d","e"], "say1": range(6)})
v2=pd.DataFrame({"anahtar":["b","c","d","e"], "say2":range(4)})
```

```python
v1
```

```python
v2
```

```python
pd.merge(v1,v2)
# kesişimlerini birleştirdi a yok
```

```python
pd.merge(v1,v2,on="anahtar")
# anahtar columnsuna göre birleştir demek
```

```python
v3=pd.DataFrame({"anahtar3":["a","b","c","c","d","e"], "say1": range(6)})
v4=pd.DataFrame({"anahtar4":["b","c","d","e"], "say2":range(4)})
pd.merge(v3,v4,left_on="anahtar3",right_on="anahtar4")
# buda farklı olanları tek tabloda birleştirme
```

```python
v3=pd.DataFrame({"anahtar3":["a","b","c","c","d","e"], "say1": range(6)})
v4=pd.DataFrame({"anahtar4":["b","c","d","e"], "say2":range(4)})
pd.merge(v3,v4,left_on="anahtar3",right_on="anahtar4",how="outer")
# kesişimlerini değil herşeyi birleştirdi diğerinde yoksa NAN eksik veri diye yazdırdı
```

```python
v3=pd.DataFrame({"anahtar3":["a","b","c","c","d","e"], "say1": range(6)})
v4=pd.DataFrame({"anahtar4":["b","c","d","e"], "say2":range(4)})
pd.merge(v3,v4,left_on="anahtar3",right_on="anahtar4",how="left")
# how=left soldakine göre oluştur
```

```python
v3=pd.DataFrame({"anahtar3":["a","b","c","c","d","e"], "say1": range(6)})
v4=pd.DataFrame({"anahtar4":["b","c","d","e"], "say2":range(4)})
pd.merge(v3,v4,left_on="anahtar3",right_on="anahtar4",how="right")
# how = right sağdakine göre oluştur NAN yok farkındaysan
```

```python
v3=pd.DataFrame({"anahtar3":["a","b","c","c","d","e"], "say1": range(6)})
v4=pd.DataFrame({"anahtar4":["b","c","d","e"], "say2":range(4)})
pd.merge(v3,v4,left_on="anahtar3",right_on="anahtar4",how="inner")
# inner yazmasak da varsayılanı kabul eder zaten ama buda kesişimleri gösterir
```

```python
v1=pd.DataFrame({"anahtar":["a","b","c","c","d","e"], "say1": range(6),
                "rakam":["bir","uc","iki","bir","bir","iki"]})
v2=pd.DataFrame({"anahtar":["b","c","d","e"], "say2":range(4),
                "rakam":["bir","bir","iki","iki"]})
```

```python
v1
```

```python
v2
```

```python
pd.merge(v1,v2,on=["anahtar","rakam"])
# kesişimi alır
```

```python
pd.merge(v1,v2,on=["anahtar","rakam"], how="outer")
```

```python
pd.merge(v1,v2,on="anahtar",how="outer")
```

```python
pd.merge(v1,v2,on="anahtar",how="outer",suffixes=("_veri1","_veri2"))
# rakam sütunlarını v1 v2 diye belirtti
```

```python
df1=pd.DataFrame({"harf":["a","a","b","b","a","c"],
                 "say":range(6)})
df2=pd.DataFrame({"deger":[3,5,7]},index=["a","b","e"])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,left_on="harf",right_index=True)
#bu komut,df1 DataFrame'ini harf sütununa göre ve df2 DataFrame'ini ise index'e göre birleştirecektir.
# right_index=True parametresi, df2 DataFrame'inin index'ine göre birleştirme yapılmasını belirtir.
# Yani, df2 DataFrame'i ile birleştirme işlemi yapılırken, df1 DataFrame'i içindeki harf sütunundaki değerlerle
# df2 DataFrame'inin index'lerinin eşleştirilmesini sağlar.
# Bu parametreyi kullanmanın nedeni, bazen birleştirme işlemi için index'in daha uygun olmasıdır. 
# Özellikle df2 DataFrame'inin index'i, birincil anahtarı veya benzersiz bir kimliği temsil ediyorsa bu kullanışlı olabilir.
```

```python
sag=pd.DataFrame([[1,2],[3,4],[5,6]],index=["a","c","d"],columns=["ali","nur"])
sol=pd.DataFrame([[7,8],[9,10],[11,12],[13,14]],index=["a","b","e","f"],columns=["efe","ata"])
```

```python
sag
```

```python
sol
```

```python
pd.merge(sag,sol,right_index=True,left_index=True,how="outer")
# hem sag hem sol indexlere göre birleştirdi
```

```python
sol.join(sag)
# sagı sola ekledi c ve d alınmadı
```

```python
sol.join(sag,how="outer")
# sagı ve sola eklerken c ve d alındı
```

```python
sol.join([sag,diger])
```

```python
veri1=pd.Series([0,1],index=["a","b"])
veri2=pd.Series([2,3,4],index=["c","d","e"])
veri3=pd.Series([5,6],index=["f","g"])
pd.concat([veri1,veri2,veri3])
```

```python
veri1=pd.Series([0,1],index=["a","b"])
veri2=pd.Series([2,3,4],index=["c","d","e"])
veri3=pd.Series([5,6],index=["f","g"])
pd.concat([veri1,veri2,veri3],axis=1,sort=False)
```

```python
veri4=pd.Series([10,11,12],index=["a","b","c"])
pd.concat([veri1,veri4],axis=1,join="inner")
# veri1 ile veri4ü concat sayesinde birleştirerek axis=1 sayesinde sütun olarak
# inner sayesinde kesişimlerini alarak yazdık
```

```python
veri4=pd.Series([10,11,12],index=["a","b","c"])
pd.concat([veri1,veri4],axis=1,join="inner").reindex(["a","b","c"])
# reindex hepsini abc sütun olarak yazdırır
```

```python
x=pd.concat([veri1,veri2,veri4],axis=1,keys=["bir","iki","uc"],sort=False)
# keys kısmı ile sütun başlıklarını belirlemedik altta anlarsın
```

```python
vs1=pd.DataFrame(np.arange(6).reshape(3,2),index=["a","b","c"],
                  columns=["bir","iki"])
vs2=pd.DataFrame(10+np.arange(4).reshape(2,2),index=["a","c"],
                columns=["uc","dort"])
pd.concat([vs1,vs2],axis=1,keys=["s1","s2"],sort=False)
```
