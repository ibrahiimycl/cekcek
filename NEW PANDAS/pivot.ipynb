```python
import pandas as pd ; import numpy as np
```

```python
nt=pd.DataFrame({"sinif":list("ABC")*4,
                "ders":("mat","fiz")*6,
                "cinsiyet":list("EKEE")*3,
                "kardes":[1,2,3]*4,
                "puan":np.arange(40,100,5)})
nt
```

```python
nt.groupby("ders")["puan"].mean()
# tabloyu derslere göre gruplandırdık sonra
# puanların ders ortalamasını yazdırdık
```

```python
nt.groupby(["ders","sinif"])["puan"].aggregate("mean")
# Bu kod, "ders" ve "sinif" sütunlarına göre gruplandırılmış olan "puan" sütununun ortalamasını hesaplar.
# Yani, her bir ders ve sınıf kombinasyonu için "puan"ın ortalamasını verir. Özetle, kodun Türkçe meali şöyle olabilir:
# "Ders" ve "sınıf" kategorilerine göre gruplandırılmış "puan"ların ortalamasını al.
```

```python
nt.pivot_table("puan",index="ders",columns="sinif")
# bu
# nt.groupby(["ders","sinif"])["puan"].aggregate("mean")
# bunun kolay hali
```

```python
nt.pivot_table(["kardes","puan"],
              index=["sinif","ders"],
              columns="cinsiyet")
```

```python
nt.pivot_table(["kardes","puan"],
              index=["sinif","ders"],
              columns="cinsiyet",margins=True)
# margins ile all diye sütun ekledik ortalamaları görüntülüyor
```

```python
nt.pivot_table(["kardes","puan"],
              index=["sinif","ders"],
              columns="cinsiyet",fill_value=0)
# NAN yerine 0 yazdırdı
```

```python
kardes=pd.cut(nt["kardes"],[0,2,3])
nt.pivot_table("puan",["ders",kardes],"sinif",fill_value=0)
```
