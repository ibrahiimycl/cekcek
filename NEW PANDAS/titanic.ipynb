```python
import pandas as pd ; import numpy as np
```

```python
s = pd.Series(range(5),index=["e","d","a","b","c"])
s
```

```python
s.sort_index() # indexe göre sıraladığım için index yazdım
```

```python
df=pd.DataFrame(np.arange(12).reshape(3,4), index=["iki","bir","üç"],
                columns=["d","a","b","c"])
df
```

```python
df.sort_index()
```

```python
df.sort_index(axis=1) # sütuna göre sıralama
```

```python
df2=pd.DataFrame({"a":[5,3,-1,9],"b":[1,-2,0,5]}) # sözlük yap. oluşturd. için süslü par { kullandık
df2
```

```python
df2.sort_values(by="b") # b süt artmasına göre sıraladı
```
